```python
import xlwt
from xlwt import Workbook

# topic_num = 2
wb = Workbook()
# book = xlsxwriter.Book('results.xlsx')
ratio_xperc_over_100perc = wb.add_sheet('ratio_xperc_over_100perc')
diff_avg_until_x = wb.add_sheet('diff_avg_until_x')
area_avg_until_x = wb.add_sheet('area_avg_until_x')
diff_at_point_of_x = wb.add_sheet('diff_at_point_of_x')
value_at_point_of_x = wb.add_sheet('value_at_point_of_x')
min_number_of_data_needed = wb.add_sheet('min_number_of_data_needed')

topic_idx = {
    '0': 0, '2': 1, '8': 2
}
topic_nums = ['0', '2', '8']


def difference(f_lc, f_rnd, n=4):
    results = []
    m = len(f_lc)
    diff = f_lc - f_rnd
    summ = 0
    index = 1
    size = 0
    for i in range(m):
        summ += diff[i]
        size += 1
        if i + 1 == int((index) * m / n):
            index += 1
            results.append(summ / size)
        #   summ = 0
    return results


def area(f, n=4):
    results = []
    m = len(f)
    # diff = f - f_rnd
    summ = 0
    index = 1
    size = 0
    for i in range(m):
        summ += f[i]
        size += 1
        if i + 1 == int((index) * m / n):
            index += 1
            results.append(summ / size)
        #   summ = 0
    return results


def difference_point(f_lc, f_rnd, n=4):
    results = []
    diff = f_lc - f_rnd
    m = len(f_lc)
    # summ = 0
    index = 1
    results.append(diff[0])
    # size = int(m/n)
    for i in range(m):
        # summ += diff[i]
        if i + 1 == int((index) * m / n):
            index += 1
            results.append(diff[i])
            # summ = 0
    return results


def at_point(f, n=4):
    results = []
    m = len(f)
    # summ = 0
    index = 1
    results.append(f[0])
    # size = int(m/n)
    for i in range(m):
        # summ += diff[i]
        if i + 1 == int((index) * m / n):
            index += 1
            results.append(f[i])
            # summ = 0
    return results


def ratio(f, n=4):
    results = []
    m = len(f)
    for i in range(n):
        index = int((i) * m / n)
        # print(index,m)
        results.append(f[index] / f[-1])
    return results


import os

arr = os.listdir('results')
files = [x for x in arr if x.endswith('.np')]
files.sort()
print(files)
# and int(x.split('-')[3])==topic_num
import numpy as np

# with open('topic_num_'+str(topic_num)+'_ratio_xperc_over_100perc.txt',mode='w',encoding='utf-8') as f:
#     f.write('Representation'+' '+'Model'+' '+'Strategy'+' '+'Metric'+' '+'0%/100% '+' '+'25%/100% '+'50%/100% '+'75%/100% '+'\n')
i = 0
i += 1
ratio_xperc_over_100perc.write(i, 0, 'representation')
ratio_xperc_over_100perc.write(i, 1, 'model')
ratio_xperc_over_100perc.write(i, 2, 'strategy')
# ratio_xperc_over_100perc.write(i, 3, 'topic')
for topic in topic_nums:
    ratio_xperc_over_100perc.write(i, 3 + 4 * topic_idx[topic], '0%/100%')
    ratio_xperc_over_100perc.write(i - 1, 3 + 4 * topic_idx[topic], 'topic:' + topic)
    ratio_xperc_over_100perc.write(i, 4 + 4 * topic_idx[topic], '25%/100%')
    ratio_xperc_over_100perc.write(i - 1, 4 + 4 * topic_idx[topic], 'topic:' + topic)
    ratio_xperc_over_100perc.write(i, 5 + 4 * topic_idx[topic], '50%/100%')
    ratio_xperc_over_100perc.write(i - 1, 5 + 4 * topic_idx[topic], 'topic:' + topic)
    ratio_xperc_over_100perc.write(i, 6 + 4 * topic_idx[topic], '75%/100%')
    ratio_xperc_over_100perc.write(i - 1, 6 + 4 * topic_idx[topic], 'topic:' + topic)
i += 1
for file in files:
    if 'f1' in file:
        print(file)
        ss = file.split('-')
        a = np.load('results/' + file)
        b = ratio(a)
        k = 3
        res = [round(b[0], k), round(b[1], k), round(b[2], k), round(b[3], k)]


        if ss[3] == '0':
            ratio_xperc_over_100perc.write(i, 0, ss[0])
            ratio_xperc_over_100perc.write(i, 1, ss[1])
            ratio_xperc_over_100perc.write(i, 2, ss[2])

        for j, rr in enumerate(res):
            ratio_xperc_over_100perc.write(i, 3 + j + 4 * topic_idx[ss[3]], rr)
        # print(b)
        if ss[3] == '8':
            i += 1


i = 0
i += 1
diff_avg_until_x.write(i, 0, 'representation')
diff_avg_until_x.write(i, 1, 'model')

for topic in topic_nums:
    diff_avg_until_x.write(i, 2 + 4 * topic_idx[topic], '0%-25%')
    diff_avg_until_x.write(i - 1, 2 + 4 * topic_idx[topic], 'topic:' + topic)
    diff_avg_until_x.write(i, 3 + 4 * topic_idx[topic], '0%-50%')
    diff_avg_until_x.write(i - 1, 3 + 4 * topic_idx[topic], 'topic:' + topic)
    diff_avg_until_x.write(i, 4 + 4 * topic_idx[topic], '0%-75%')
    diff_avg_until_x.write(i - 1, 4 + 4 * topic_idx[topic], 'topic:' + topic)
    diff_avg_until_x.write(i, 5 + 4 * topic_idx[topic], '0%-100%')
    diff_avg_until_x.write(i - 1, 5 + 4 * topic_idx[topic], 'topic:' + topic)

i += 1
for file in files:
    ss = file.split('-')
    if 'f1' in file and ss[2] == 'lc':
        a = np.load('results/' + file)
        cand = [h for h in files if file.replace('-lc-', '-Random-') in h and 'f1' in file and h.split('-')[3] == ss[
            3] and 'baseline' not in file]
        print(file)

        r = np.load('results/' + cand[0])

        b = difference(a, r)
        k = 3
        # print(b)
        res = [round(b[0], k), round(b[1], k), round(b[2], k), round(b[3], k)]

        if ss[3] == '0':
            diff_avg_until_x.write(i, 0, ss[0])
            diff_avg_until_x.write(i, 1, ss[1])

        for j, rr in enumerate(res):
            diff_avg_until_x.write(i, 2 + j + 4 * topic_idx[ss[3]], rr)
        # print(b)

        if ss[3] == '8':
            i += 1

i = 0
i += 1

area_avg_until_x.write(i, 0, 'representation')
area_avg_until_x.write(i, 1, 'model')
area_avg_until_x.write(i, 2, 'strategy')
for topic in topic_nums:

    area_avg_until_x.write(i, 3 + 4 * topic_idx[topic], '0%-25%')
    area_avg_until_x.write(i - 1, 3 + 4 * topic_idx[topic], 'topic:' + topic)
    area_avg_until_x.write(i, 4 + 4 * topic_idx[topic], '0%-50%')
    area_avg_until_x.write(i - 1, 4 + 4 * topic_idx[topic], 'topic:' + topic)
    area_avg_until_x.write(i, 5 + 4 * topic_idx[topic], '0%-75%')
    area_avg_until_x.write(i - 1, 5 + 4 * topic_idx[topic], 'topic:' + topic)
    area_avg_until_x.write(i, 6 + 4 * topic_idx[topic], '0%-100%')
    area_avg_until_x.write(i - 1, 6 + 4 * topic_idx[topic], 'topic:' + topic)

i += 1
for file in files:
    ss = file.split('-')
    if 'f1' in file:
        print(file)
        a = np.load('results/' + file)

        b = area(a)
        k = 3

        res = [round(b[0], k), round(b[1], k), round(b[2], k), round(b[3], k)]


        if ss[3] == '0':
            area_avg_until_x.write(i, 0, ss[0])
            area_avg_until_x.write(i, 1, ss[1])
            area_avg_until_x.write(i, 2, ss[2])

        for j, rr in enumerate(res):
            area_avg_until_x.write(i, 3 + j + 4 * topic_idx[ss[3]], rr)

        if ss[3] == '8':
            i += 1


i = 0
i += 1

diff_at_point_of_x.write(i, 0, 'representation')
diff_at_point_of_x.write(i, 1, 'model')

for topic in topic_nums:
    diff_at_point_of_x.write(i, 2 + 5 * topic_idx[topic], '0%')
    diff_at_point_of_x.write(i - 1, 2 + 5 * topic_idx[topic], 'topic:' + topic)
    diff_at_point_of_x.write(i, 3 + 5 * topic_idx[topic], '25%')
    diff_at_point_of_x.write(i - 1, 3 + 5 * topic_idx[topic], 'topic:' + topic)
    diff_at_point_of_x.write(i, 4 + 5 * topic_idx[topic], '50%')
    diff_at_point_of_x.write(i - 1, 4 + 5 * topic_idx[topic], 'topic:' + topic)
    diff_at_point_of_x.write(i, 5 + 5 * topic_idx[topic], '75%')
    diff_at_point_of_x.write(i - 1, 5 + 5 * topic_idx[topic], 'topic:' + topic)
    diff_at_point_of_x.write(i, 6 + 5 * topic_idx[topic], '100%')
    diff_at_point_of_x.write(i - 1, 6 + 5 * topic_idx[topic], 'topic:' + topic)

i += 1
for file in files:
    ss = file.split('-')
    if 'f1' in file and ss[2] == 'lc':
        print(file)
        a = np.load('results/' + file)
        cand = [h for h in files if
                file.replace('-lc-', '-Random-') in h and h.split('-')[3] == ss[3] and 'baseline' not in file]
        print(file, cand[0])
        r = np.load('results/' + cand[0])
        b = difference_point(a, r)
        k = 3

        res = [round(b[0], k), round(b[1], k), round(b[2], k),
               round(b[3], k), round(b[4], k)]


        if ss[3] == '0':
            diff_at_point_of_x.write(i, 0, ss[0])
            diff_at_point_of_x.write(i, 1, ss[1])

        for j, rr in enumerate(res):
            diff_at_point_of_x.write(i, 2 + j + 5 * topic_idx[ss[3]], rr)
        # print(b)
        if ss[3] == '8':
            i += 1


i = 0
i += 1

value_at_point_of_x.write(i, 0, 'representation')
value_at_point_of_x.write(i, 1, 'model')
value_at_point_of_x.write(i, 2, 'strategy')

for topic in topic_nums:
    value_at_point_of_x.write(i, 3 + 5 * topic_idx[topic], '0%')
    value_at_point_of_x.write(i - 1, 3 + 5 * topic_idx[topic], 'topic:' + topic)
    value_at_point_of_x.write(i, 4 + 5 * topic_idx[topic], '25%')
    value_at_point_of_x.write(i - 1, 4 + 5 * topic_idx[topic], 'topic:' + topic)
    value_at_point_of_x.write(i, 5 + 5 * topic_idx[topic], '50%')
    value_at_point_of_x.write(i - 1, 5 + 5 * topic_idx[topic], 'topic:' + topic)
    value_at_point_of_x.write(i, 6 + 5 * topic_idx[topic], '75%')
    value_at_point_of_x.write(i - 1, 6 + 5 * topic_idx[topic], 'topic:' + topic)
    value_at_point_of_x.write(i, 7 + 5 * topic_idx[topic], '100%')
    value_at_point_of_x.write(i - 1, 7 + 5 * topic_idx[topic], 'topic:' + topic)

i += 1

for file in files:
    if 'f1' in file:
        print(file)
        ss = file.split('-')

        a = np.load('results/' + file)

        b = at_point(a)
        k = 3

        res = [round(b[0], k), round(b[1], k), round(b[2], k),
               round(b[3], k), round(b[4], k)]

        if ss[3] == '0':
            value_at_point_of_x.write(i, 0, ss[0])
            value_at_point_of_x.write(i, 1, ss[1])
            value_at_point_of_x.write(i, 2, ss[2])

        for j, rr in enumerate(res):
            value_at_point_of_x.write(i, 3 + j + 5 * topic_idx[ss[3]], rr)

        if ss[3] == '8':
            i += 1



def min_data(f, err=0.01):
    m = len(f)
    maximum = max(f)
    min_data = 20
    at_max = 0
    for i in range(m):
        at_max = f[i]
        if f[i] >= maximum - err:
            break
        min_data += 50
    return [min_data, at_max, maximum]


i = 0
i += 1

min_number_of_data_needed.write(i, 0, 'representation')
min_number_of_data_needed.write(i, 1, 'model')
min_number_of_data_needed.write(i, 2, 'strategy')

for topic in topic_nums:
    min_number_of_data_needed.write(i, 3 + 3 * topic_idx[topic], 'min number of data')
    min_number_of_data_needed.write(i - 1, 3 + 3 * topic_idx[topic], 'topic:' + topic)
    min_number_of_data_needed.write(i, 4 + 3 * topic_idx[topic], 'at point of min data')
    min_number_of_data_needed.write(i - 1, 4 + 3 * topic_idx[topic], 'topic:' + topic)
    min_number_of_data_needed.write(i, 5 + 3 * topic_idx[topic], 'max achieved')
    min_number_of_data_needed.write(i - 1, 5 + 3 * topic_idx[topic], 'topic:' + topic)

i += 1

for file in files:
    if 'f1' in file:
        ss = file.split('-')

        a = np.load('results/' + file)

        b = min_data(a)
        k = 3

        res = [round(b[0], k), round(b[1], k), round(b[2], k)]

        if ss[3] == '0':
            min_number_of_data_needed.write(i, 0, ss[0])
            min_number_of_data_needed.write(i, 1, ss[1])
            min_number_of_data_needed.write(i, 2, ss[2])

        for j, rr in enumerate(res):
            min_number_of_data_needed.write(i, 3 + j + 3 * topic_idx[ss[3]], rr)

        if ss[3] == '8':
            i += 1



wb.save('results.xls')
```

# Make Plots

## Figures for each Model or Embedding

### Setting
One figure for each keyword, all experiments with specified keyword wil be in the figure

```python

keywords = ['LR', 'svm', 'mlp', 'rf', 'Ada', 'lc', 'Random', 'BERT', 'GLOVE', 'tweetBERT', 'gp', 'lda', 'qda', 'knn3',
            'knn5']

# keywords = ['GLOVE100','GLOVE50','GLOVE25']


measures = ['f1']
```

### Plot

```python


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re
from matplotlib.pyplot import figure

# Apply the default theme
sns.set_style("darkgrid")
markers = ['.', 'o', 'v', '^', '<', '8', 's', 'p', '*'] * 100
for topic_num in topic_nums:
    for measure in measures:
        for keyword in keywords:
            random_added = False
            majority_added = False
            figure(figsize=(5, 5))
            index = 0
            for file in files:
                if measure in file and keyword in file and ('t' + keyword not in file) and file.split('-')[
                    3] == topic_num:
                    print('plot:', file)
                    a = np.load('results/' + file)
                    d = {measure: a, 'percentage of data': [i * (100 / (len(a) - 1)) for i in range(len(a))]}
                    pdnumsqr = pd.DataFrame(d)
                    string = file[:-3]
                    string = string.replace(keyword + '-', '')
                    string = string.replace('Random', 'RND')
                    string = string.replace('f1', '')
                    string = string.replace('accuracy', '')
                    string = string.replace('tweetBERT', 'TBERT')
                    string = string.replace('batch_lc', 'BLC')
                    string = re.sub(rf'-(\d)-', rf'', string).upper()
                    # string = re.sub(rf'-(\d)-.*',rf'-topic\1',string).upper()
                    # print(string)
                    sns.lineplot(x='percentage of data', y=measure, data=pdnumsqr, label=string, marker=markers[index])
                    index += 1
                    # plt.plot(a, label=file[:-3])
                # plt.ylabel(measure)
                if 'baseline' in file and keyword in file and ('t' + keyword not in file) and file.split('-')[
                    3] == topic_num:
                    print('plot:', file)
                    a = np.load('results/' + file)

                    if 'majority' not in file:
                        if not random_added:
                            plt.axhline(y=a, label='random')
                            random_added = True

                    elif not majority_added:
                        d = {measure: a, 'percentage of data': [i * (100 / (len(a) - 1)) for i in range(len(a))]}
                        pdnumsqr = pd.DataFrame(d)
                        string = file[:-3]
                        string = string.replace(keyword + '-', '')
                        string = string.replace('Random', 'RND')
                        string = string.replace('f1', '')
                        string = string.replace('baseline', '')
                        string = string.replace('accuracy', '')

                        string = re.sub(rf'-(\d)-', rf'', string).upper()
                        # string = re.sub(rf'-(\d)-.*',rf'-topic\1',string).upper()
                        # print(string)
                        sns.lineplot(x='percentage of data', y=measure, data=pdnumsqr, label='majority',
                                     marker=markers[index])

                        index += 1
                        majority_added = True
            plt.legend(loc='lower right', fancybox=True, shadow=True, ncol=1, fontsize=10)
            plt.title(keyword.upper() + ' - topic:' + str(topic_num))
            plt.savefig('plots/' + keyword + '-' + str(topic_num) + '-' + measure + '.png')
            plt.clf()

```
